```python
import json
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
```

```python
df = pd.read_csv('imdb_top_1000.csv', index_col='Series_Title')
df = df.dropna(subset=['Gross'])
df['Gross'] = df['Gross'].apply(lambda x: int(x.replace(",", "")))
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df
```

```python
#print(X_test.dtypes)
X=df[['Released_Year','IMDB_Rating','No_of_Votes']]
y=df['Gross']
```

```python
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
for depth in range(1, 11):
    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(X_train, y_train)  
    test_pred = clf.predict(X_test)  
   
    test_correct = test_pred == y_test.values  
    test_accuracy = np.mean(test_correct)  

    print(f'max_depth={depth} | Test_Accuracy: {test_accuracy:.2f}')
    
```

```python
for depth in range(1, 11):
    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(X_train, y_train)  
    train_pred = clf.predict(X_train)  
   
    train_correct = train_pred == y_train.values  
    train_accuracy = np.mean(train_correct)  

    print(f'max_depth={depth} | Train_Accuracy: {train_accuracy:.2f}')
```

```python

test_pred = clf.predict(X_test)
results = X_test.copy()
results['Actual'] = y_test
results['Predicted'] = test_pred
wrong_preds = results[results['Actual'] != results['Predicted']]
for idx, row in wrong_preds.head(5).iterrows():
    print(f"\n {idx}")
    for col in X_test.columns:
        print(f"  {col}: {row[col]}")
    print(f"\n  Actual: {row['Actual']}")
    print(f"  Predicted: {row['Predicted']}")
```
